```python
import torch 
from torchvision.transforms import ToTensor
import matplotlib.pyplot as plt
from PIL import Image

from DDPM import DDPM
from UNet import UNet
```

```python
ddpm = DDPM(UNet(), device="mps")
```

```python
image = Image.open('poke.jpg')
image
```

```python
image_t = ToTensor()(image).unsqueeze(0)
image_t.shape
```

```python
noised = ddpm.forward(image_t.to("mps"), torch.tensor(0).to('mps'))
```

```python
plt.imshow(noised.squeeze(0).permute(1, 2, 0).cpu())
```
